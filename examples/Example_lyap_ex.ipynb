```python
import dill
import numpy as np
```

After running example_lyap_ex.py we get as output a pkl file which can be analyzed

```python
with open('Lor63_N2000.pkl', 'rb') as reservoir:
    lor63_N2000 = dill.load(reservoir)
```

```python
lor63_N2000.ifsave = False
```

```python
lor63_N2000.predict(15, acc = 3)
```

Here we can plot the global lyapunov exponents of the predicting reservoir. Red dashes are the lyapunov exponents of the input system while blue lines are the LE of the reservoir.  The predicting reservoir should approximately reproduce the exponents of the system.

```python
lor63_N2000.plotLE(10)
```
